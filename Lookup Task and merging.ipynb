```python
df1=pd.read_excel('Q2_Input.xlsx',sheet_name='First_Record')
df2=pd.read_excel('Q2_Input.xlsx',sheet_name='Second_Record')
df3=pd.read_excel('Q2_Input.xlsx',sheet_name='Third_Record')
```

```python
lookup_data=pd.read_csv("Q2_Lookup.csv")
```

```python
final=pd.concat([df1,df2,df3])
```

```python
df=final[final['Company']=='Acuity']
```

```python
#Lookup 
final_out=pd.merge(df,lookup_data,left_on='Company',right_on='Name')
final_out
```

```python
final1=final_out.groupby(['Year','Company','Code','Rating'])['New Joinees'].sum().reset_index()
```
